```python
data = pd.read_csv("/data/thejustinh/TidyQuotes.csv")
data.head()
```

Generate a 2D-List of Stock Quotes to be graphed.

```python
x = 0
y = len(data['Date'])
data['Date2'] = data['Date'].map(lambda d: mdates.date2num(dt.datetime.strptime(d, "%Y-%m-%d")))
ohlc = []
while x < y:
    ohlc.append([data['Date2'][x], data['Open'][x], data['High'][x], data['Low'][x], data['Close'][x]])
    x += 1
```

```python
assert len(ohlc) == len(data)
```

## Dow Jones Indistrial Average Candlestick Charts

```python
def graph_quotes(data = ohlc):
    fig, ax = plt.subplots()
    ax.xaxis_date()
    plt.xlabel("Date")
    plt.ylabel("Price")
    candlestick_ohlc(ax, data, width =0.08, colorup='g', colordown='r')
    plt.title("Plot of DJIA Prices over time")
    plt.grid()
    plt.show()
```

```python
graph_quotes(ohlc)
```

While manually reading each headline may be helpful to get a grasp of what happened during a specific day, it isn't efficient at all. Below, I perform sentiment analysis which analysis each headline and giving a polarity score for a headline's neutrality, positivity, and negativity.

## Is there a relationship between events and price changes?
### Sentiment Analysis
Here we peform sentiment analysis on each headline giving a polarity score for neutrality, negativity, and positivity. A compound polarity score is given as the aggregate score.

```python
sid = SentimentIntensityAnalyzer()
neu = []
neg = []
compound = []
pos = []
for index, row in news.iterrows():
    ss = sid.polarity_scores(str(row))
    for k in ss:
        if k == 'compound':
            compound.append(ss[k])
        if k == 'neu':
            neu.append(ss[k])
        if k == 'pos':
            pos.append(ss[k])
        if k == 'neg':
            neg.append(ss[k])
```

```python
assert len(neu) == len(neg) == len(compound) == len(pos)
```

```python
news['Compound'] = compound
news['Negative'] = neg
news['Positive'] = pos
news['Neutral'] = neu
```

```python
news.head().loc[:, 'Compound':'Neutral']
```

```python
status = []
for x in data.Stock_change:
    if x > 0:
        status.append(1)
    else:
        status.append(0)
data['Status_change'] = status
```

```python
sentiment_data = pd.merge(data, news)
sentiment_data.to_csv("/data/thejustinh/DataSentiment.csv")
```

### Price Change vs. Sentiment

```python
pvs = sentiment_data[['Stock_change', 'Oil_Change', 'Gas_Change', 'Gold_Change', 
                      'Compound', 'Negative', 'Neutral', 'Positive']]
```

```python
def plot_price_vs_sentiment(market, sentiment):
    plt.scatter(sentiment, market, data=pvs)
    plt.title("Price Change vs. Sentiment")
    plt.xlabel("Sentiment Score")
    plt.ylabel("Price Change")
    plt.show()
```

```python
interact(plot_price_vs_sentiment, market = {'DJIA':'Stock_change', 'Crude Oil':'Oil_Change', 
                                            'Natural Gas':'Gas_Change'}, 
         sentiment = {'Compound':'Compound', 'Negative':'Negative', 'Neutral':'Neutral', 'Positive':'Positive'})
```

This interact graph shows us something very interesting. For the sentiments of positivity, negativity, and neutrality, there seems to be no correlation between the sentiments and the market. However, if we look of the graphs of any market vs. the compound score of a sentiment, then we get a much more meaningful graph. 

Any knowledgable investor would say that an important aspect of trading is to minimize risks whenever possible. From the graph above, we can see that a compound sentiment score greater than -0.25 would mean that your returns would range from -500 points to 250 points - a higher compound sentiment score would reduce this threshold. Because the market is so volatile, it is really hard to say if tomorrow's market will increase or decrease, but by looking at the compound sentiments of news headlines, we can minimize our risks and improve our chances of losing less.
